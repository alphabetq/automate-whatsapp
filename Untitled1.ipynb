```python
from selenium import webdriver
from bs4 import BeautifulSoup
import requests
from time import sleep
import os
import urllib.request
from selenium.webdriver.common.keys import Keys 
from selenium.webdriver.common.action_chains import ActionChains
from datetime import datetime
import random
from collections import deque
import paho.mqtt.client as mqtt
import time,sys,itertools
import threading
import socket 

hostname = socket.gethostname()    
IPAddr = socket.gethostbyname(hostname)
instruction_queue = deque()
is_connected = False
client = None

dirpath = os.getcwd()
browser = webdriver.Chrome(executable_path = dirpath+"/chromedriver")
url = "https://news.google.com/topstories?hl=en-MY&gl=MY&ceid=MY:en"
root = "https://news.google.com"
browser.get('https://web.whatsapp.com')

bot_users = {"File Transfer":"File Transfer"}
head_count_list = {}
commands = {'deactivate':'deactivate', 'activate':'activate', 'news':'news', 'help':'help'}

def mqtt_broadcast(client, msg):
    client.publish("selenium", msg, qos=0, retain=False)

def create_client(client, is_connected):
    client = mqtt.Client(f"user")

    # client callbacks
    def on_connect(client, userdata, flags, rc):
        """ The callback for when the client receives a CONNACK response from the server."""

        print(f"Connected with result code {str(rc)}")
        is_connected = True

    def on_message(client, userdata, msg):
        """The callback for when a PUBLISH message is received from the server."""
        try:

            decoded_msg = msg.payload.decode("utf-8")

            instruction_queue.append(decoded_msg)


        except Exception as e:
            print(e)

    def on_disconnect(client, userdata, rc):
        """The callback for when disconnected from the broker."""

        print(f"Disconnected with result code {str(rc)}")
        client.loop_stop()
        print("Loop stopped.")
        is_connected = False

    client.on_connect = on_connect
    client.on_message = on_message
    client.on_disconnect = on_disconnect
    
    return client, is_connected


client, is_connected = create_client(client, is_connected)
client.connect(IPAddr)
client.loop_start()
client.subscribe("selenium")

while True:
    try:
        text_box = browser.find_element_by_xpath("//div[contains(@class, '_3F6QL _2WovP')]/div[2]")
        name = browser.find_element_by_class_name("_3XrHh").text  # Conversation name
        message = browser.find_elements_by_class_name("_3zb-j")[-1]  # the message content
        
        def send_msg(msg):
            return(f'*Jarvis:* {msg}\n')

        def tiny_url(url):
            apiurl = "http://tinyurl.com/api-create.php?url="
            tinyurl = urllib.request.urlopen(apiurl + url).read()
            return tinyurl.decode("utf-8")

        def getNews():
            text_box.send_keys(send_msg(f"Retrieving top 5 news from {root}, please wait..."))
            soup = BeautifulSoup(requests.get(url).content, "html5lib")
            articles = soup.find_all('article', class_="MQsxIb xTewfe R7GTQ keNKEd j7vNaf Cc0Z5d EjqUne")
            caption = [i.find_all('h3',class_="ipQwMb ekueJc gEATFF RD0gLb")[0].text for i in articles[:5]]
            # news = [i.find_all('a',class_="ipQwMb ekueJc gEATFF RD0gLb")[0].text for i in articles[:5]]
            links = [root+i.find('a')['href'][1:] for i in articles[:5]]
            links = [tiny_url(link) for link in links]
            for i in range(5):
                text_box.send_keys(f"{i+1}.) {caption[i]} {links[i]}\n")
                
                
        while len(instruction_queue) != 0:
            queue = instruction_queue.popleft()
            if '@jarvis' not in queue:
                text_box.send_keys(send_msg(queue))
            
        
        if '@jarvis' in message.text.lower() and 'jarvis:' not in message.text.lower():
            if name in bot_users:
                
                if commands['deactivate'] in message.text.lower():
                    del bot_users[name]
                    text_box.send_keys(send_msg("*Jarvis has left the chat.*"))
                elif commands['activate'] in message.text.lower():
                    text_box.send_keys(send_msg("*Already signed in, bitch.*"))
                    
                elif commands['news'] in message.text.lower():
                    getNews()
                elif commands['help'] in message.text.lower():
                    text_box.send_keys(send_msg(f"Available commands: {commands.keys()}"))
                elif 'whos your daddy' in message.text.lower():
                    text_box.send_keys(send_msg(f"Ronn."))
                elif 'thank' in message.text.lower():
                    reply = ['no prob', 'ma nigga', 'ur welcome', ';)']
                    text_box.send_keys(send_msg(random.choice(reply)))
                else:
                    text_box.send_keys(send_msg(f"Type '@jarvis' and the following commands: {commands.keys()}"))
                    
            elif commands['deactivate'] in message.text.lower():
                text_box.send_keys(send_msg("*Already signed off, bitch.*"))
            elif commands['activate'] in message.text.lower():
                bot_users.update({name:name})
                text_box.send_keys(send_msg("*Activated. Ready for service.*"))
                
    except Exception as e:
        print(e)
        pass
    sleep(2) # A 2 second pause so that the program doesn't run too fast
```
